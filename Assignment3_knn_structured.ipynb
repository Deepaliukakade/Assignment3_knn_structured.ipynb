Assignment‑3: k‑Nearest Neighbour (kNN) — Iris Dataset

Objective: Implement k‑Nearest Neighbour algorithm using Python ML libraries and analyze predictions.

```python
# Assignment-3: k-Nearest Neighbour (kNN) — Iris Dataset

# 🔹 Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

print("Libraries imported.")

# 🔹 Step 2: Load Dataset from UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

columns = ["sepal_length","sepal_width","petal_length","petal_width","class"]

df = pd.read_csv(url, names=columns)

print("\nFirst 5 rows of dataset:")
print(df.head())

# 🔹 Step 3: Dataset Overview
print("\nShape:", df.shape)
print("\nColumn Names:", df.columns.tolist())

print("\nPreview of dataset:")
print(df.head())

# 🔹 Step 4: Data Types & Summary
print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# 🔹 Step 5: Check Missing Values
print("\nMissing Values in each column:")
print(df.isnull().sum())

# 🔹 Step 6: Handle Missing Values (if found)

# Fill numeric columns with mean
for col in df.select_dtypes(include=['float64','int64']).columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill categorical columns with mode
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing Values After Handling:")
print(df.isnull().sum())

# 🔹 Step 7: Remove Duplicate Records
before = df.shape[0]

df.drop_duplicates(inplace=True)

after = df.shape[0]

print("\nDuplicates removed:", before - after)

# 🔹 Step 8: Encode Categorical Column
encoder = LabelEncoder()

df["class"] = encoder.fit_transform(df["class"])

print("\nDataset After Encoding:")
print(df.head())

# 🔹 Step 9: Feature Scaling
scaler = StandardScaler()

cols = ["sepal_length","sepal_width","petal_length","petal_width"]

df[cols] = scaler.fit_transform(df[cols])

print("\nDataset After Feature Scaling:")
print(df.head())

# 🔹 Step 10: Save Final Dataset
df.to_csv("assignment2_cleaned_iris.csv", index=False)

print("\nSaved: assignment2_cleaned_iris.csv")
```

✔ Conclusion

In this assignment, the Iris dataset was successfully preprocessed using Python. The dataset was analyzed for structure, data types, missing values, and duplicate records. Missing values were handled, categorical data was encoded into numeric form, and feature scaling was applied to normalize the dataset. Finally, the cleaned dataset was saved for further machine learning tasks such as applying the k-Nearest Neighbour (kNN) algorithm. These preprocessing steps help improve the quality and performance of machine learning models.
